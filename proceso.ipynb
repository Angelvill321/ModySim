```python
mapping_dict = {
    "Nott'ham Forest": 'Nottingham Forest',
    'Brighton': 'Brighton and Hove Albion',
    'Wolves': 'Wolverhampton Wanderers',
    'Newcastle Utd': 'Newcastle United',
    'West Brom': 'West Bromwich Albion',
    'West Ham': 'West Ham United',
    'Sheffield Utd': 'Sheffield United',
    'Manchester Utd': 'Manchester United',
    'Tottenham': 'Tottenham Hotspur'
}

df['opponent_normalized'] = df['opponent'].replace(mapping_dict, regex=False)

set_team_final = set(df['team'].unique())
set_opponent_final = set(df['opponent_normalized'].unique())

print(f"VerificaciÃ³n Final: {set_team_final == set_opponent_final}")
```

```python
import numpy as np

df['season_year'] = df['season'].str.split('/').str[0].astype(int)
max_id = df['season_year'].max()

decay_rate = 0.5 # Tasa de decaimiento. Un valor mÃ¡s alto hace que los partidos viejos cuenten MUCHO menos.

df['weight'] = np.exp(-decay_rate * (max_id - df['season_year']))

df_modelo_local = pd.DataFrame({
    'Equipo': df['team'],
    'Oponente': df['opponent'],
    'Goles_FAVOR': df['gf'],
    'Goles_CONTRA': df['ga'],
    'Es_Local': 1,
    'season_year': df['season_year'],
    'weight': df['weight']
})

df_modelo_visitante = pd.DataFrame({
    'Equipo': df['opponent'],
    'Oponente': df['team'],
    'Goles_FAVOR': df['ga'],
    'Goles_CONTRA': df['gf'],
    'Es_Local': 0,
    'season_year': df['season_year'],
    'weight': df['weight']
})

df_modelo = pd.concat([df_modelo_local, df_modelo_visitante], ignore_index=True)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# FÃ³rmula del modelo: Goles_FAVOR se explica por Ventaja de LocalÃ­a,
# Fuerza de Ataque del Equipo y Fuerza de Defensa del Oponente.
formula = 'Goles_FAVOR ~ Es_Local + C(Equipo) + C(Oponente)'

poisson_model_weighted = smf.glm(
    formula=formula, 
    data=df_modelo, 
    family=sm.families.Poisson()
).fit(freq_weights=df_modelo['weight'])

print("âœ… Modelo de Poisson ponderado ajustado exitosamente.")
```

```python

def extraer_fuerzas(modelo, df_equipos):
    params = modelo.params
    intercepto = params['Intercept']
    ventaja_local = params['Es_Local']
    fuerzas = {'intercepto': intercepto, 'ventaja_local': ventaja_local}
    equipos = df_equipos['Equipo'].unique()

    for equipo in equipos:
        fuerzas[equipo] = {'ataque': 0.0, 'defensa': 0.0}
        ataque_key = f'C(Equipo)[T.{equipo}]'
        fuerzas[equipo]['ataque'] = params.get(ataque_key, 0.0)
        defensa_key = f'C(Oponente)[T.{equipo}]'
        fuerzas[equipo]['defensa'] = params.get(defensa_key, 0.0)

    return fuerzas

fuerzas_estimadas = extraer_fuerzas(poisson_model_weighted, df_modelo)
print("Fuerzas base extraÃ­das.")

debutantes_sin_historial = ['Ipswich Town']
penalizacion_fuerza = -1

descendidos = ['Burnley', 'Sheffield United', 'Luton Town']
ascendidos = ['Leicester City', 'Southampton', 'Ipswich Town']
df_ultima_temp = df_modelo[df_modelo['weight'] == 1.0].copy()
equipos_2024 = set(df_ultima_temp['Equipo'].unique())

equipos_para_simulacion = equipos_2024 - set(descendidos)
equipos_para_simulacion.update(ascendidos)
TODOS_LOS_EQUIPOS_SIMULACION = list(equipos_para_simulacion)

#AGREGA los debutantes al diccionario antes de penalizar (SOLUCIONA EL PROBLEMA)
for equipo in debutantes_sin_historial:
     if equipo not in fuerzas_estimadas:
         # AÃ±adir al diccionario con fuerza inicial 0.0 (fuerza promedio de la liga)
         fuerzas_estimadas[equipo] = {'ataque': 0.0, 'defensa': 0.0}
         print(f"Importante: '{equipo}' agregado al diccionario de fuerzas.")

#APLICA la penalizaciÃ³n extrema a los debutantes
for equipo in debutantes_sin_historial:
     fuerzas_estimadas[equipo]['ataque'] = penalizacion_fuerza
     fuerzas_estimadas[equipo]['defensa'] = penalizacion_fuerza
     print(f"FUERZA PENALIZADA: {equipo} ajustada a {penalizacion_fuerza}")


partidos = []
for local in TODOS_LOS_EQUIPOS_SIMULACION:
    for visitante in TODOS_LOS_EQUIPOS_SIMULACION:
        if local != visitante:
            partidos.append({'local': local, 'visitante': visitante})
df_fixture = pd.DataFrame(partidos)
```

```python

def simular_partido_montecarlo(equipo_local, equipo_visitante, fuerzas_dict):
    f = fuerzas_dict 
    
    ataque_local = f.get(equipo_local, {'ataque': 0.0})['ataque']
    defensa_local_reciproca = f.get(equipo_local, {'defensa': 0.0})['defensa']
    ataque_visitante = f.get(equipo_visitante, {'ataque': 0.0})['ataque']
    defensa_visitante_reciproca = f.get(equipo_visitante, {'defensa': 0.0})['defensa']

    log_lambda_local = (f['intercepto'] + f['ventaja_local'] + ataque_local + defensa_visitante_reciproca)
    lambda_local = np.exp(log_lambda_local)
    log_lambda_visitante = (f['intercepto'] + ataque_visitante + defensa_local_reciproca)
    lambda_visitante = np.exp(log_lambda_visitante)
    
    goles_local = np.random.poisson(lam=lambda_local, size=1)[0]
    goles_visitante = np.random.poisson(lam=lambda_visitante, size=1)[0]
    
    return goles_local, goles_visitante

def inicializar_tabla_optimizada(equipos):
    tabla = {equipo: {'Puntos': 0, 'Victorias': 0, 'Empates': 0, 'Derrotas': 0, 'GF': 0, 'GC': 0, 'GD': 0} for equipo in equipos} 
    return tabla


def actualizar_tabla_optimizada(tabla, local, visitante, gl, gv):
    tabla[local]['GF'] += gl
    tabla[local]['GC'] += gv
    tabla[visitante]['GF'] += gv
    tabla[visitante]['GC'] += gl
    
    if gl > gv:
        tabla[local]['Puntos'] += 3; tabla[local]['Victorias'] += 1; tabla[visitante]['Derrotas'] += 1
    elif gl < gv:
        tabla[visitante]['Puntos'] += 3; tabla[visitante]['Victorias'] += 1; tabla[local]['Derrotas'] += 1
    else:
        tabla[local]['Puntos'] += 1; tabla[visitante]['Puntos'] += 1
        tabla[local]['Empates'] += 1; tabla[visitante]['Empates'] += 1
        
    return tabla

num_torneos_a_simular = 10000 
victorias_por_equipo = {equipo: 0 for equipo in TODOS_LOS_EQUIPOS_SIMULACION}
todos_los_rankings = [] 

semilla_fija = 120 
np.random.seed(semilla_fija)
print(f"\nIniciando simulaciÃ³n de {num_torneos_a_simular} torneos con semilla {semilla_fija}...")

for i in range(num_torneos_a_simular):
    
    tabla_actual_dict = inicializar_tabla_optimizada(TODOS_LOS_EQUIPOS_SIMULACION)
    
    for index, row in df_fixture.iterrows():
        local = row['local']
        visitante = row['visitante']
        goles_local, goles_visitante = simular_partido_montecarlo(local, visitante, fuerzas_estimadas)
        tabla_actual_dict = actualizar_tabla_optimizada(tabla_actual_dict, local, visitante, goles_local, goles_visitante)
        
    tabla_actual = pd.DataFrame.from_dict(tabla_actual_dict, orient='index')
    tabla_actual['GD'] = tabla_actual['GF'] - tabla_actual['GC']
    tabla_actual = tabla_actual.sort_values(by=['Puntos', 'GD'], ascending=False)
    tabla_actual['Rank'] = np.arange(1, len(tabla_actual) + 1)
    
    campeon = tabla_actual.index[0]
    victorias_por_equipo[campeon] += 1
    
    rank_df = tabla_actual[['Puntos', 'GD', 'Rank']].reset_index().rename(columns={'index': 'Equipo'})
    todos_los_rankings.append(rank_df)
    
    if (i + 1) % 1000 == 0:
        print(f"SimulaciÃ³n {i + 1}/{num_torneos_a_simular} completada.")

# --- RESULTADOS FINALES Y EXPORTACIÃ“N ---
probabilidad_campeon = {equipo: victorias / num_torneos_a_simular for equipo, victorias in victorias_por_equipo.items()}
ranking_final = sorted(probabilidad_campeon.items(), key=lambda item: item[1], reverse=True)

print("\n--- ðŸ† RANKING DE PROBABILIDAD DE CAMPEONATO FINAL ---")
for equipo, prob in ranking_final[:10]: 
    print(f"**{equipo}**: {prob:.2%}")

# Exportar Resumen de Probabilidades por PosiciÃ³n
df_resultados_completos = pd.concat(todos_los_rankings, ignore_index=True)
df_probabilidades_posicion = (
    df_resultados_completos.groupby(['Equipo', 'Rank']).size().reset_index(name='Frecuencia')
)
df_probabilidades_posicion['Probabilidad'] = df_probabilidades_posicion['Frecuencia'] / num_torneos_a_simular
df_resumen_probabilidades = df_probabilidades_posicion.pivot_table(
    index='Equipo', columns='Rank', values='Probabilidad', fill_value=0
)

df_resultados_completos.to_csv('resultados_simulacion_completa.csv', index=False)
print("âœ… Resultados completos (Puntos y Rank por simulaciÃ³n) exportados a 'resultados_simulacion_completa.csv'")

df_resumen_probabilidades.to_csv('resumen_probabilidades_por_posicion.csv')
print("\nâœ… Resumen de probabilidades de posiciÃ³n exportado a 'resumen_probabilidades_por_posicion.csv'")
```
